## Descriptive Analysis

### Objective

The objective of this task is to perform descriptive statistical analysis on the restaurant dataset.

The analysis includes:

- Understanding numerical variables
- Calculating statistical measures
- Exploring categorical variables
- Identifying the most popular cuisines
- Identifying cities with the highest number of restaurants
- Drawing meaningful business insights from the data

```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

pd.set_option("display.max_columns",None)
```

```python
# Load cleaned dataset

df = pd.read_csv("/Cleaned_Dataset.csv")

print("Dataset Loaded Successfully")
```

```python
df.head()
```

```python
print("Rows :",df.shape[0])
print("Columns :",df.shape[1])
```

```python
numerical_columns = df.select_dtypes(include=np.number)

numerical_columns.columns
```

```python
numerical_columns.describe().T
```

```python
mean_values = numerical_columns.mean()

mean_values
```

```python
median_values = numerical_columns.median()

median_values
```

```python
mode_values = numerical_columns.mode().iloc[0]

mode_values
```

```python
std_values = numerical_columns.std()

std_values
```

```python
variance = numerical_columns.var()

variance
```

```python
minimum = numerical_columns.min()

minimum
```

```python
maximum = numerical_columns.max()

maximum
```

```python
range_values = maximum-minimum

range_values
```

```python
statistics = pd.DataFrame({

'Mean':mean_values,

'Median':median_values,

'Mode':mode_values,

'Standard Deviation':std_values,

'Variance':variance,

'Minimum':minimum,

'Maximum':maximum,

'Range':range_values

})

statistics.round(2)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Aggregate rating"],bins=20,kde=True)

plt.title("Distribution of Aggregate Rating")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Votes"],bins=30,color="green")

plt.title("Distribution of Votes")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Average Cost for two"],bins=30,color="orange")

plt.title("Average Cost for Two")

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df["Aggregate rating"])

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df["Votes"])

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df["Average Cost for two"])

plt.show()
```

```python
country = df["Country Code"].value_counts()

country
```

```python
plt.figure(figsize=(10,6))

sns.countplot(y="Country Code",data=df,order=df["Country Code"].value_counts().index)

plt.title("Restaurant Distribution by Country")

plt.show()
```

```python
city = df["City"].value_counts()

city.head(20)
```

```python
top10_city = df["City"].value_counts().head(10)

top10_city
```

```python
plt.figure(figsize=(12,6))

sns.barplot(

x=top10_city.values,

y=top10_city.index

)

plt.title("Top 10 Cities with Highest Number of Restaurants")

plt.xlabel("Number of Restaurants")

plt.ylabel("City")

plt.show()
```

```python
cuisine = df["Cuisines"].value_counts()

cuisine.head(20)
```

```python
top10_cuisine = df["Cuisines"].value_counts().head(10)

top10_cuisine
```

```python
plt.figure(figsize=(12,6))

sns.barplot(

x=top10_cuisine.values,

y=top10_cuisine.index

)

plt.title("Top 10 Most Popular Cuisines")

plt.xlabel("Number of Restaurants")

plt.ylabel("Cuisine")

plt.show()
```

```python
city_percentage = (df["City"].value_counts(normalize=True)*100).round(2)

city_percentage.head(10)
```

```python
cuisine_percentage = (df["Cuisines"].value_counts(normalize=True)*100).round(2)

cuisine_percentage.head(10)
```
