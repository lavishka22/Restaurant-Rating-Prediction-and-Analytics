```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("/Feature_Engineered_Dataset.csv")

print("Dataset Loaded Successfully")
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Aggregate rating"],
    bins=20,
    kde=True
)

plt.title("Distribution of Aggregate Ratings")

plt.xlabel("Aggregate Rating")

plt.ylabel("Frequency")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Votes"],
    bins=40,
    kde=True
)

plt.title("Distribution of Customer Votes")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Average Cost for two"],
    bins=40
)

plt.title("Distribution of Average Cost for Two")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="Price Category"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Restaurant Distribution by Price Category")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Price Category",
    y="Aggregate rating"
)

plt.title("Aggregate Rating by Price Category")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Votes",
    y="Aggregate rating",
    alpha=0.6
)

plt.title("Votes vs Aggregate Rating")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Average Cost for two"],
    df["Aggregate rating"],
    s=df["Votes"]/5,
    alpha=0.5
)

plt.xlabel("Average Cost for Two")

plt.ylabel("Aggregate Rating")

plt.title("Bubble Chart: Cost vs Rating vs Votes")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
pairplot_columns = [
    "Aggregate rating",
    "Votes",
    "Average Cost for two",
    "Cuisine Count",
    "Cost Per Person"
]

sns.pairplot(df[pairplot_columns])

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(
    data=df,
    x="Has Online delivery",
    y="Aggregate rating"
)

plt.title("Aggregate Rating by Online Delivery")

plt.tight_layout()

plt.show()
```

```python
top_city = df["City"].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_city.values,
    y=top_city.index
)

plt.title("Top 10 Cities by Restaurant Count")

plt.tight_layout()

plt.show()
```

```python
top_cuisine = df["Cuisines"].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_cuisine.values,
    y=top_cuisine.index
)

plt.title("Top 10 Cuisines")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(
    data=df,
    x="Has Table booking",
    y="Aggregate rating"
)

plt.title("Aggregate Rating by Table Booking")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(
    data=df,
    x="Has Online delivery",
    y="Aggregate rating"
)

plt.title("Aggregate Rating by Online Delivery")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Price Category",
    y="Votes"
)

plt.title("Customer Votes by Price Category")

plt.tight_layout()

plt.show()
```

```python
country_mapping = {
    1: "India",
    14: "Australia",
    30: "Brazil",
    37: "Canada",
    94: "Indonesia",
    148: "New Zealand",
    162: "Philippines",
    166: "Qatar",
    184: "Singapore",
    189: "South Africa",
    191: "Sri Lanka",
    208: "Turkey",
    214: "United Arab Emirates",
    215: "United Kingdom",
    216: "United States"
}

df["Country"] = df["Country Code"].map(country_mapping)

print("Country column created successfully.")
```

```python
df[["Country Code", "Country"]].head()
```

```python
highest_rated_cuisine = (
    df.groupby("Cuisines")
    .agg(
        Restaurant_Count=("Restaurant ID", "count"),
        Average_Rating=("Aggregate rating", "mean")
    )
)

highest_rated_cuisine = highest_rated_cuisine[
    highest_rated_cuisine["Restaurant_Count"] >= 10
]

highest_rated_cuisine_name = (
    highest_rated_cuisine["Average_Rating"].idxmax()
)

dashboard = pd.DataFrame({
    "Metric": [
        "Total Restaurants",
        "Countries",
        "Cities",
        "Most Popular Cuisine",
        "Highest Rated Cuisine",
        "Best Model",
        "Best R² Score"
    ],

    "Value": [
        len(df),
        df["Country"].nunique(),
        df["City"].nunique(),
        df["Cuisines"].mode()[0],
        highest_rated_cuisine_name,
        "Random Forest",
        0.971
    ]
})

dashboard
```
